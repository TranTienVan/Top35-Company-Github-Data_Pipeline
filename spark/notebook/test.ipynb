```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
```

```python
r = requests.get("https://github.com/orgs/IBM/repositories")


soup = BeautifulSoup(r.text, "html.parser")
```

```python
datetime.now() - timedelta(hours = 1)
```

```python
repos = soup.find_all("li", {"class": "Box-row"})

for repo in repos:
    name = repo.find("a", {"class": "d-inline-block"})
    time = repo.find("relative-time", {"class": "no-wrap"})
    
    if datetime.strptime(time["datetime"].replace("T", " ")[:-1], "%Y-%m-%d %H:%M:%S") + timedelta(hours = 7)> datetime.now() - timedelta(hours = 1):
        print(name.getText(strip = True), time["datetime"].replace("T", " ")[:-1])
```
